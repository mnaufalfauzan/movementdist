```python
import os
import shutil

# Path ke folder utama yang berisi sub-folder
main_folder = r"D:\Downloads\DATA MOVEMENT DISTRIBUTION"
output_folder = r"D:\Downloads\DATA MOVEMENT DISTRIBUTION 2"

# Membuat folder output jika belum ada
os.makedirs(output_folder, exist_ok=True)

# Loop melalui semua sub-folder
for root, dirs, files in os.walk(main_folder):
    for file in files:
        # Periksa apakah file memiliki ekstensi .csv
        if file.lower().endswith(".csv"):
            file_path = os.path.join(root, file)  # Path lengkap file asli
            output_path = os.path.join(output_folder, file)  # Path tujuan file
            
            try:
                # Menyalin file ke folder tujuan
                shutil.copy(file_path, output_path)
                print(f"Copied: {file_path} -> {output_path}")
            except Exception as e:
                print(f"Error copying {file}: {e}")

print("All files copied successfully!")
```

```python
import os
import pandas as pd

# Path ke folder yang berisi semua file harian
input_folder = r"D:\Downloads\DATA MOVEMENT DISTRIBUTION 2"  # Folder dengan file harian
output_folder = r"D:\Downloads\DATA MOVEMENT DISTRIBUTION IDN"  # Folder untuk hasil filter

# Membuat folder output jika belum ada
os.makedirs(output_folder, exist_ok=True)

# Loop untuk memproses setiap file CSV
for file in os.listdir(input_folder):
    if file.endswith(".csv"):  # Memastikan hanya file CSV yang diproses
        file_path = os.path.join(input_folder, file)
        
        try:
            # Membaca file CSV
            data = pd.read_csv(file_path)
            
            # Memfilter data untuk country == "IDN"
            filtered_data = data[data['country'] == 'IDN']
            
            if not filtered_data.empty:  # Jika ada data IDN, simpan hasilnya
                output_file = os.path.join(output_folder, file)
                filtered_data.to_csv(output_file, index=False)
                print(f"Filtered data saved to {output_file}")
        except Exception as e:
            print(f"Error processing file {file}: {e}")

print("All filtering complete!")

# Path ke folder hasil
print(f"Filtered files are saved in: {output_folder}")
```

```python
import os
import pandas as pd

# Path ke folder yang berisi semua file harian
input_folder = r"D:\Downloads\DATA MOVEMENT DISTRIBUTION 2"  # Folder dengan file harian
output_folder = r"D:\Downloads\DATA MOVEMENT DISTRIBUTION IDN"  # Folder untuk hasil filter

# Membuat folder output jika belum ada
os.makedirs(output_folder, exist_ok=True)

# Loop untuk memproses setiap file CSV
for file in os.listdir(input_folder):
    if file.endswith(".csv"):  # Memastikan hanya file CSV yang diproses
        file_path = os.path.join(input_folder, file)
        
        try:
            # Membaca file CSV
            data = pd.read_csv(file_path)
            
            # Memfilter data untuk country == "IDN"
            filtered_data = data[data['country'] == 'IDN']
            
            if not filtered_data.empty:  # Jika ada data IDN, tambahkan kolom tanggal
                # Extract tanggal dari nama file (bagian setelah "_", sebelum ".csv")
                date = file.split('_')[1].replace('.csv', '')
                filtered_data['date'] = date  # Tambahkan kolom tanggal
                
                # Simpan hasil ke folder output
                output_file = os.path.join(output_folder, file)
                filtered_data.to_csv(output_file, index=False)
                print(f"Filtered data with date saved to {output_file}")
        except Exception as e:
            print(f"Error processing file {file}: {e}")

print("All filtering complete!")

# Path ke folder hasil
print(f"Filtered files with dates are saved in: {output_folder}")
```

```python
import os
import pandas as pd

# Path ke folder hasil filter
input_folder = r"D:\Downloads\DATA MOVEMENT DISTRIBUTION IDN"  # Folder dengan file hasil filter
output_folder = r"D:\Downloads\DATA MOVEMENT DISTRIBUTION IDN COMBINED"  # Folder untuk file gabungan
os.makedirs(output_folder, exist_ok=True)  # Membuat folder jika belum ada

# Path lengkap untuk file gabungan
output_file = os.path.join(output_folder, "Combined_IDN_Data.csv")

# DataFrame kosong untuk menggabungkan semua file
combined_data = pd.DataFrame()

# Loop untuk membaca semua file CSV di folder input
for file in os.listdir(input_folder):
    if file.endswith(".csv"):  # Memastikan hanya file CSV yang diproses
        file_path = os.path.join(input_folder, file)
        
        try:
            # Membaca file CSV
            data = pd.read_csv(file_path)
            
            # Menambahkan data ke DataFrame gabungan
            combined_data = pd.concat([combined_data, data], ignore_index=True)
        except Exception as e:
            print(f"Error processing file {file}: {e}")

# Menyimpan DataFrame gabungan ke file CSV
combined_data.to_csv(output_file, index=False)
print(f"Combined data saved to {output_file}")
```

```python
import os
import pandas as pd

# Path ke folder hasil filter
input_folder = r"D:\Downloads\DATA MOVEMENT DISTRIBUTION IDN"  # Folder dengan file hasil filter
output_folder = r"D:\Downloads\DATA MOVEMENT DISTRIBUTION IDN COMBINED"  # Folder untuk file gabungan
os.makedirs(output_folder, exist_ok=True)  # Membuat folder jika belum ada

# Path lengkap untuk file gabungan
output_file = os.path.join(output_folder, "Combined_IDN_Data.csv")

# DataFrame kosong untuk menggabungkan semua file
combined_data = pd.DataFrame()

# Tambahkan log untuk memeriksa file input
print(f"Reading files from: {input_folder}")

# Loop untuk membaca semua file CSV di folder input
for file in os.listdir(input_folder):
    if file.endswith(".csv"):  # Memastikan hanya file CSV yang diproses
        print(f"Processing file: {file}")
        file_path = os.path.join(input_folder, file)
        
        try:
            # Membaca file CSV
            data = pd.read_csv(file_path)
            print(f"File {file} read successfully with {len(data)} rows.")
            
            # Menambahkan data ke DataFrame gabungan
            combined_data = pd.concat([combined_data, data], ignore_index=True)
        except Exception as e:
            print(f"Error processing file {file}: {e}")

# Menyimpan DataFrame gabungan ke file CSV
if not combined_data.empty:
    combined_data.to_csv(output_file, index=False)
    print(f"Combined data saved to {output_file}")
else:
    print("No data combined. Please check the input files.")
```
